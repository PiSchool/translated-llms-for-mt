```python
from sentence_transformers import SentenceTransformer
import pandas as pd
```

```python
import torch
```

```python
embedder = SentenceTransformer("paraphrase-multilingual-mpnet-base-v2")
```

**Embedding of Translated prompting dataset**

You'll have to upload manually the data that you want to embed.

```python
translated_it_en = pd.read_csv('/content/translated-big-it-en-dev.csv')
translated_it_es = pd.read_csv('/content/translated-big-it-es-dev.csv')
```

```python
embedded_it_en_IT = embedder.encode(translated_it_en["source"].tolist(), convert_to_tensor=True).cpu()
embedded_it_en_EN = embedder.encode(translated_it_en["target"].tolist(), convert_to_tensor=True).cpu()
embedded_it_es_IT = embedder.encode(translated_it_es["source"].tolist(), convert_to_tensor=True).cpu()
embedded_it_es_ES = embedder.encode(translated_it_es["target"].tolist(), convert_to_tensor=True).cpu()
```

```python
torch.save(embedded_it_en_IT, "./translated-big-it-en-dev-embedding_IT.pt")
torch.save(embedded_it_en_EN, "./translated-big-it-en-dev-embedding_EN.pt")
torch.save(embedded_it_es_IT, "./translated-big-it-es-dev-embedding_IT.pt")
torch.save(embedded_it_es_ES, "./translated-big-it-es-dev-embedding_ES.pt")
```

**Embedding of Flores prompting dataset**

```python
with open('/content/spa_Latn.dev') as f:
    flores_es = f.readlines()
with open('/content/ita_Latn.dev') as f:
    flores_it = f.readlines()
with open('/content/eng_Latn.dev') as f:
    flores_en = f.readlines()
```

```python
embedded_it = embedder.encode(flores_it, convert_to_tensor=True).cpu()
embedded_en = embedder.encode(flores_en, convert_to_tensor=True).cpu()
embedded_es = embedder.encode(flores_es, convert_to_tensor=True).cpu()
```

```python
torch.save(embedded_it, "./flores-dev-embedding_it.pt")
torch.save(embedded_en, "./flores-dev-embedding_en.pt")
torch.save(embedded_es, "./flores-dev-embedding_es.pt")
```
